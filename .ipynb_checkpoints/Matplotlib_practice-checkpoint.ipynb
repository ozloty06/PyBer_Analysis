```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to al ist of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated each month.
y_axis = [10.02, 23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis, marker="+", color="blue", linewidth=0.5, label='Boston')
# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.subplots(figsize=(20,10))
# Set the y limit between 0 and 45
plt.ylim(0,45)
plt.xlim(0,5)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig= plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add a legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(y_axis, x_axis)
```

```python
# Create a horizontal bar chart
plt.barh(x_axis,y_axis, color = "magenta", label='Boston')
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
# Invert y-axis to show dates from Jan - Dec rather than other way around
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to al ist of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated each month.
y_axis = [10.02, 23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(y_axis,x_axis, color='red')
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.title("PyBer Fare by Month")
# Invert y-axis to show dates from Jan - Dec rather than other way around
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*8)
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
explode_values = (0,0,0,0,0,0,0.3,0,0,0,0,0)

plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,labels=x_axis,autopct='%.1f%%')

plt.show()
```

```python
# Assign 12 colors, one for each month
colors=["slateblue","magenta", "lightblue", "green", "yellowgreen","greenyellow","yellow","orange", "gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0,0,0,0,0.3,0,0,0,0,0)

plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```
